```python
# IMPORTATION DES BIBLIOTHEQUES
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# CHARGEMENT DE DONNEES
data = pd.read_csv('Base_de_donne_iris.csv')
print(data)

#charger les données en ligne
url = ('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
data_online = pd.read_csv(url)

# afficher les données en ligne
print(data_online)


# classifier les X et les Y

X = data_online.drop('species', axis=1)
Y = data_online['species']
# afficher les X et les Y
# varibale de sortie 
print(X)
# la cible
print(Y)

# methode 2 pour classifier les X et les Y

# ========== Deuxième méthode : sélection par position avec iloc ==========

# On ne garde que les colonnes numériques + species pour rester cohérent
# (ou on travaille directement sur data_online)
X = data_online.iloc[:, :-1]   # toutes les colonnes sauf la dernière
Y = data_online.iloc[:, -1]    # dernière colonne = species

print("\n=== Deuxième méthode (iloc) ===")
print("X (5 premières lignes) :")
print(X)
print("\nY (5 premières lignes) :")
print(Y)

# DIVISER LES DONNEES EN ENSEMBLE D'ENTRAINNEMENT ET DE TEST
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# afficher les dimensions des ensembles d'entraînement et de test
print("\nDimensions de X_train :", X_train.shape)  
print("Dimensions de Y_train :", Y_train.shape)
print("Dimensions de X_test :", X_test.shape)
print("Dimensions de Y_test :", Y_test.shape

      
!pip install scikit-learn pandas numpy matplotlib
# importer les bibliothèques nécessaires

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report


# charger les données stockées en local 
#data = pd.read_csv('base.csv')
#print(data)

#charger les données en ligne
url = ('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
data_online = pd.read_csv(url)

# afficher les données en ligne
#print(data_online)

# classifier les X et les Y

#X = data_online.drop('species', axis=1)
#Y = data_online['species']
# afficher les X et les Y
# varibale de sortie 
#print(X)
# la cible
#print(Y)

# ========== Deuxième méthode : sélection par position avec iloc ==========

# On ne garde que les colonnes numériques + species pour rester cohérent
# (ou on travaille directement sur data_online)
X = data_online.iloc[:, :-1]   # toutes les colonnes sauf la dernière
Y = data_online.iloc[:, -1]    # dernière colonne = species

print("\n=== Deuxième méthode (iloc) ===")
print("X (toutes les lignes) :")
print(X)
print("\nY (toutes les lignes) :")
print(Y)

# ========== Division en ensemble d'apprentissage et de test ==========

# Séparation : 80% train, 20% test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("\n=== Division Train / Test ===")
print("X_train shape :", X_train.shape)
print("X_test shape  :", X_test.shape)
print("Y_train shape :", Y_train.shape)
print("Y_test shape  :", Y_test.shape)

print("\nX_train (toutes les lignes) :")
print(X_train)
print("\nY_train (toutes les lignes) :")
print(Y_train)

# ========== Création du modèle d'arbre de décision ==========

# 1. Créer le modèle
arbre = DecisionTreeClassifier(random_state=42)

# 2. Entraîner le modèle sur les données d'apprentissage
arbre.fit(X_train, Y_train)

# 3. Prédire sur l'ensemble de test
Y_pred = arbre.predict(X_test)

# 4. Évaluer le modèle
accuracy = accuracy_score(Y_test, Y_pred)

print("\n=== Modèle Arbre de Décision ===")
print("Prédictions sur X_test :")
print(Y_pred)
print("\nVraies valeurs Y_test :")
print(Y_test.values)
print("\nPrécision (accuracy) :", accuracy)
print("\nRapport de classification :")
print(classification_report(Y_test, Y_pred))

# Installation si nécessaire
!pip install scikit-learn matplotlib graphviz

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns

# 1. CHARGEMENT DES DONNÉES
data = pd.read_csv('Base_de_donne_iris.csv')

# Ajouter les noms d'espèces
data['nom_espece'] = data['species'].map({
    0: 'Iris setosa', 
    1: 'Iris versicolor', 
    2: 'Iris virginica'
})

print("📊 Aperçu des données :")
print(data.head())
print(f"\nDimensions : {data.shape}")

# 2. PRÉPARATION DES DONNÉES
X = data[['petal_length', 'petal_width']]
Y = data['species']
noms_classes = ['Setosa', 'Versicolor', 'Virginica']

# 3. DIVISION TRAIN/TEST
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=42, stratify=Y
)

print(f"\n📈 Dimensions des ensembles :")
print(f"Entraînement : {X_train.shape[0]} échantillons")
print(f"Test : {X_test.shape[0]} échantillons")

# 4. CRÉATION DE L'ARBRE DE DÉCISION
model = DecisionTreeClassifier(
    max_depth=3,           # Profondeur maximale
    min_samples_split=5,   # Échantillons min pour diviser
    min_samples_leaf=2,    # Échantillons min par feuille
    criterion='gini',      # Critère de division ('gini' ou 'entropy')
    random_state=42
)

# Entraînement
model.fit(X_train, Y_train)

# 5. ÉVALUATION
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)

print("\n📊 PERFORMANCES DU MODÈLE :")
print(f"Précision entraînement : {accuracy_score(Y_train, Y_train_pred):.2%}")
print(f"Précision test : {accuracy_score(Y_test, Y_test_pred):.2%}")

print("\n📋 Rapport de classification détaillé :")
print(classification_report(Y_test, Y_test_pred, target_names=noms_classes))

# 6. MATRICE DE CONFUSION
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Matrice de confusion
cm = confusion_matrix(Y_test, Y_test_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=noms_classes, 
            yticklabels=noms_classes,
            ax=axes[0])
axes[0].set_title('Matrice de confusion')
axes[0].set_ylabel('Vraie classe')
axes[0].set_xlabel('Classe prédite')

# 7. VISUALISATION DE L'ARBRE DE DÉCISION
plot_tree(model, 
          feature_names=['Longueur pétale', 'Largeur pétale'],
          class_names=noms_classes,
          filled=True, 
          rounded=True, 
          fontsize=10,
          ax=axes[1])
axes[1].set_title('Arbre de décision')

plt.tight_layout()
plt.show()

# 8. IMPORTANCE DES CARACTÉRISTIQUES
importances = model.feature_importances_
features = ['Longueur pétale', 'Largeur pétale']

plt.figure(figsize=(8, 5))
plt.bar(features, importances, color=['blue', 'green'])
plt.title('Importance des caractéristiques')
plt.ylabel('Importance')
for i, v in enumerate(importances):
    plt.text(i, v + 0.01, f'{v:.2%}', ha='center')
plt.show()

# 9. VISUALISATION DES FRONTIÈRES DE DÉCISION
plt.figure(figsize=(10, 7))

# Créer une grille
x_min, x_max = X['petal_length'].min() - 0.5, X['petal_length'].max() + 0.5
y_min, y_max = X['petal_width'].min() - 0.5, X['petal_width'].max() + 0.5
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

# Prédictions sur la grille
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Affichage
plt.contourf(xx, yy, Z, alpha=0.4, cmap='RdYlBu', levels=[-0.5, 0.5, 1.5, 2.5])
scatter = plt.scatter(X['petal_length'], X['petal_width'], 
                     c=Y, cmap='RdYlBu', edgecolors='black', s=50)

# Ajouter les noms
for i, txt in enumerate(data['nom_espece'].unique()):
    subset = data[data['nom_espece'] == txt]
    plt.scatter(subset['petal_length'].mean(), subset['petal_width'].mean(), 
               marker='x', s=200, c='black', linewidths=2)

plt.xlabel('Longueur du pétale (cm)')
plt.ylabel('Largeur du pétale (cm)')
plt.title('Frontières de décision de l\'arbre de décision')
plt.colorbar(scatter, ticks=[0, 1, 2], label='Espèce')
plt.show()

# 10. RÈGLES DE DÉCISION
print("\n🌳 RÈGLES DE DÉCISION DE L'ARBRE :")
from sklearn.tree import export_text
regles = export_text(model, feature_names=['Longueur pétale', 'Largeur pétale'])
print(regles)

# 11. TESTER AVEC DE NOUVELLES FLEURS
print("\n🌸 TEST AVEC DE NOUVELLES FLEURS :")
nouvelles_fleurs = pd.DataFrame({
    'petal_length': [1.5, 4.5, 5.5],
    'petal_width': [0.3, 1.5, 2.2]
})

predictions = model.predict(nouvelles_fleurs)
for i, (longueur, largeur, pred) in enumerate(zip(
    nouvelles_fleurs['petal_length'], 
    nouvelles_fleurs['petal_width'], 
    predictions)):
    print(f"Fleur {i+1} ({longueur}cm, {largeur}cm) → {noms_classes[pred]}")
```

```python
# =============================================================================
# CELLULE 4 : PRÉPARATION DES DONNÉES POUR LE MACHINE LEARNING
# =============================================================================

print("=" * 60)
print("🔧 PRÉPARATION DES DONNÉES POUR LE MACHINE LEARNING")
print("=" * 60)

# Séparation features/target
X = data_complete[numeric_cols].copy()
Y = data_complete['species'].copy()

print(f"✅ Features (X) : {X.shape[1]} caractéristiques")
print(f"   {list(X.columns)}")
print(f"✅ Target (Y)   : {Y.nunique()} classes")
print(f"   {list(Y.unique())}")

# Encodage des labels
le = LabelEncoder()
Y_encoded = le.fit_transform(Y)

print(f"\n📌 Encodage des espèces :")
for i, nom in enumerate(le.classes_):
    print(f"   {i} → {nom}")

# Division train/test avec stratification
X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y_encoded, 
    test_size=0.2, 
    random_state=42, 
    stratify=Y_encoded
)

print(f"\n📊 Division des données :")
print(f"   • Train : {X_train.shape[0]} échantillons ({X_train.shape[0]/len(X)*100:.0f}%)")
print(f"   • Test  : {X_test.shape[0]} échantillons ({X_test.shape[0]/len(X)*100:.0f}%)")

# Vérification de la distribution
print(f"\n📈 Distribution des classes :")
for i, nom in enumerate(['Setosa', 'Versicolor', 'Virginica']):
    train_pct = (Y_train == i).sum() / len(Y_train) * 100
    test_pct = (Y_test == i).sum() / len(Y_test) * 100
    print(f"   {nom:12} → Train: {train_pct:.1f}% | Test: {test_pct:.1f}%")

print("\n✅ Données prêtes pour l'entraînement !")
```
